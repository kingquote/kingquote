```python
import random as rd
import pylab as py

n = 500 # number of agents

class agent:
    pass

def initialise():
    global agents  
    agents = []
    for i in range(n):
        ag = agent()
        ag.x = rd.uniform(-1,1) # select x and y values: random number between (-1,1)
        ag.y = rd.uniform(-1,1) 
        ag.newx = 0 # # we need to store two positions: current and previous 
        ag.newy = 0
        agents.append(ag)
    
def observe():
    global agents
    py.cla()   
    py.plot([ag.x for ag in agents], [ag.y for ag in agents], 'bo')
    py.axis('image')
    py.axis([-50, 50, -50, 50])
  
def update():
    global agents
    Cx = py.mean([ag.x for ag in agents]) # centre of swarm in x and y
    Cy = py.mean([ag.y for ag in agents]) 
    
    for ag in agents:
        ag.newx = 2*ag.x - ag.newx + 0.2*(Cx - ag.x)/py.norm(Cx - ag.x) 
        # 2ag is ag + variation in ag (=ag - agnew) 
        # norm removes minus sign; term is a restoring force
        ag.newy = 2*ag.y - ag.newy + 0.2*(Cy - ag.y)/py.norm(Cy - ag.y) 
        # now we need to update ag.x, that becomes 
        # the calculated ag.new; and then ag.new 
        # becomes the old ag.x (which is stored for next round)
        ag.x, ag.newx = ag.newx, ag.x 
        ag.y, ag.newy = ag.newy, ag.y
    

import pycxsimulator
pycxsimulator.GUI().start(func=[initialise, observe, update])

```

Exercise 2

The code is very simple but already presents swarm-like behaviour.
Which are the obvious limitations?

    2-D, Overlapping entities, real entity is not likely to know the location of the center of the swarm, only the immideate neighbors are visible.

How could you modify the code to make sure the swarm never flies away from the plot?? 

    Limit the possible values in the location updating, eg through if statements, or maybe floor/ceilings? Example below. I added some randomness as well,
    since multiple entities that are in the same spot will move to the same new spot in the deterministic model.

What could make the swarm simulation more realistic? Anything birds do that current rules donâ€™t represent? How could changes be implemented?

    Movement of the swarm: we could take the movement vector of the center of the swarm since the last step and make each bird move in the same
    direction by adding this vector or a fraction of it to the existing movement vector.
    I did this below, with 1.8 instead of 2 to avoid the swarm speeding away.


```python
import random as rd
import pylab as py
import statistics as stat

n = 500 # number of agents
lim = 100 # limits of plot

class agent:
    pass

def initialise():
    global agents  
    agents = []
    global Cxold
    Cxold = 0
    global Cyold
    Cyold = 0
    for i in range(n):
        ag = agent()
        ag.x = rd.uniform(-1,1) # select x and y values: random number between (-1,1)
        ag.y = rd.uniform(-1,1) 
        ag.newx = 0 # # we need to store two positions: current and previous 
        ag.newy = 0
        agents.append(ag)
    
def observe():
    global agents
    global Cxold
    global Cyold
    py.cla()   
    py.plot([ag.x for ag in agents], [ag.y for ag in agents], 'bo')
    py.plot(Cxold, Cyold, 'ro')
    py.axis('image')
    py.axis([-lim, lim, -lim, lim])
  
def update():
    global agents
    global Cxold
    global Cyold
    Cx = py.mean([ag.x for ag in agents]) # centre of swarm in x and y
    Cy = py.mean([ag.y for ag in agents]) 
    
    for ag in agents:
        ag.newx = 2*ag.x - ag.newx + \
            0.7*(Cx - ag.x)/py.norm(Cx - ag.x) + \
            0.4 * stat.median([-0.5, 0.01*(1*Cx - Cxold) + \
                    0.05*(0 - Cx)/py.norm(0 - Cx), 0.5]) +\
            rd.uniform(-0.1,0.1)
        # 2ag is ag + variation in ag (=ag - agnew) 
        # norm removes minus sign; term is a restoring force
        ag.newy = 2*ag.y - ag.newy + \
            0.7*(Cy - ag.y)/py.norm(Cy - ag.y) + \
            0.4 * stat.median([-0.5, 0.01*(1*Cy - Cyold) + \
                    0.05*(0 - Cy)/py.norm(0 - Cy), 0.5]) + \
            rd.uniform(-0.1,0.1)

        if ag.newx > lim:
            ag.newx = lim
        if ag.newx < -lim:
            ag.newx = -lim
        if ag.newy > lim:
            ag.newy = lim
        if ag.newy < -lim:
            ag.newy = -lim
        
        # now we need to update ag.x, that becomes 
        # the calculated ag.new; and then ag.new 
        # becomes the old ag.x (which is stored for next round)
        ag.x, ag.newx = ag.newx, ag.x 
        ag.y, ag.newy = ag.newy, ag.y
    
    Cxold, Cyold = Cx, Cy
    

import pycxsimulator
pycxsimulator.GUI().start(func=[initialise, observe, update])

```
